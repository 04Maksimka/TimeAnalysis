```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots

def dft_power_spectrum(signal: np.ndarray, sample_rate: float = 1.0):
    """
    Compute the one-sided power spectrum of a real signal.

    Parameters
    ----------
    signal      : 1-D array of real values
    sample_rate : samples per unit time (e.g. 1.0 = one sample per day)

    Returns
    -------
    freqs : array of frequencies (cycles per unit time)
    power : array of power values
    """
    N = len(signal)
    X = np.fft.rfft(signal)          # rfft exploits real-valued symmetry → N/2+1 bins
    freqs = np.fft.rfftfreq(N, d=1 / sample_rate)
    power = (np.abs(X) ** 2) / N
    return freqs, power
```

```python
rng = np.random.default_rng(42)
N, fs = 512, 1.0          # 512 trading days, 1 sample/day

t = np.arange(N) / fs
signal = (
    np.sin(2 * np.pi * t / 20)   # 20-day cycle
    + 0.5 * np.sin(2 * np.pi * t / 5)  # 5-day cycle
    + 0.3 * rng.standard_normal(N)      # noise
)

freqs, power = dft_power_spectrum(signal, sample_rate=fs)

fig = make_subplots(rows=2, cols=1,
                    subplot_titles=("Synthetic price signal", "Power spectrum"))

fig.add_trace(go.Scatter(x=t, y=signal, mode="lines", name="signal",
                         line=dict(color="#4361ee", width=1)), row=1, col=1)

fig.add_trace(go.Scatter(x=1 / freqs[1:], y=power[1:], mode="lines",
                         name="power", line=dict(color="#f72585")), row=2, col=1)

fig.update_xaxes(title_text="Trading day", row=1)
fig.update_xaxes(title_text="Period (days)", row=2, type="log")
fig.update_yaxes(title_text="Price", row=1)
fig.update_yaxes(title_text="Power", row=2)
fig.update_layout(height=520, showlegend=False, template="simple_white")
fig.show()
```
